## Line Chart


A line chart displays information as a series of data points or markers, Line charts work out of the box with matplotlib. You can have multiple lines in a line chart, change color, change type of line and much more

let's take an examlpe to understand and translat the idea of line chart 

we neet to display/plot the velovity of 2 objects , let's do it 

```python
## firt , create time list 
time = [0 ,5,10,15,20,25,30]
## create velocity of the first object 
velocity_obj_1 = [10,15,20,20,30,15,0]
## create velocity of the second object 
velocity_obj_2 = [10,12,15,12,20,10,0]
```

```python
## we use plot function to drow line chart 
plt.plot(time  , velocity_obj_1)
## we created velocity of the first by the time 
```

```python
## we use plot function to drow line chart 
plt.plot(time , velocity_obj_2)
## we created velocity of the second by the time 
```

```python
## let's merge the 2 velocity of 2 objects by using plt.plot 2 times 
plt.plot(time  , velocity_obj_1)
plt.plot(time  , velocity_obj_2)
```

#### [<matplotlib.lines.Line2D at 0x1aaeba93af0>] , we can remove this line by put ; at the end of last line ,  or using show function 

```python
plt.plot(time , velocity_obj_1) ; 
```

```python
plt.plot(time , velocity_obj_1) 
plt.show()
```

```python
# we can put title for out graph by using plt.title() , and customize the size of font by using fontsize attribute 
plt.plot(time  , velocity_obj_1)
plt.plot(time  , velocity_obj_2)

# Title
plt.title("velocity of 2 objects " , fontsize = 20 )

plt.show()
```

```python
# legend() function 
# A legend is an area describing the elements of the graph. 
plt.plot(time  , velocity_obj_1)
plt.plot(time  , velocity_obj_2)

# In the matplotlib library, there's a function called legend() which is used to Place a legend on the axesplt
plt.legend(["object 1 " , "object 2 "]) ; 
```

```python
# make titles for x and y axis
plt.plot(time  , velocity_obj_1)
plt.plot(time  , velocity_obj_2)

plt.xlabel("Time" , fontsize = 12 )
plt.ylabel("velocity of objects" , fontsize = 12 ) ; 

plt.legend(["object 1 " , "object 2 "]) ; 
```

```python
# let's make our graph more beautiful and clear
# try to run 2 codes/ ways to know diffrent 

plt.plot(time  , velocity_obj_1 , linewidth = 3 ) # make width of line more thick 
plt.plot(time  , velocity_obj_2 , linewidth = 3 ) # make width of line more thick 


# plt.plot(time  , velocity_obj_1 ,'*', linewidth = 3 , markersize =12   )
# plt.plot(time  , velocity_obj_2 , '*', linewidth = 3 , markersize =12  ); 

plt.legend(["object 1 " , "object 2 "]) ; 
```

How to make our graph bigger? 
we use a function called figure to make the graph bigger 

```python
# graph size 
plt.figure(figsize=(15 , 10 )) # 10 is the width of graph  , y is the hight of graph 

# plot / drow graph 
plt.plot(time  , velocity_obj_1)
plt.plot(time  , velocity_obj_2)

# Title
plt.title("velocity of 2 objects " , fontsize = 20 )

# legend
plt.legend(["object 1 " , "object 2 "]) 

# titles 
plt.xlabel("Time" , fontsize = 15 )
plt.ylabel("velocity of objects" , fontsize = 15 ) 

# show our graph and remove result of drawing result 
plt.show()
```

## Scatter Plot


In a scatter plot, the values of 2 variables are plotted as points on a 2-dimensional grid. Additonally, you can also use a third variable to determine the size or color of the points.
let's do previous example with scatter plot 

```python
# use plt.scatter to create a scatter graph
# c refers to coller of points 
# s refers to size of points  
# alpha refers to color opacity of each point , number of alpha must between 0.1 and 1  
plt.scatter(time , velocity_obj_1 , c= 'r' , s = 100 , alpha= 0.8)
plt.scatter(time , velocity_obj_2 , c= 'k' , s = 100 , alpha= 0.9)
# you can use grid() function to get grid lines
# plt.grid()
```

## Histogram and Frequency Distribution


A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

For example, lets say that you have the following data about the age of 100 individuals:



```python
age = [1,1,2,3,3,5,7,8,9,10,
10,11,11,13,13,15,16,17,18,18,
18,19,20,21,21,23,24,24,25,25,
25,25,26,26,26,27,27,27,27,27,
29,30,30,31,33,34,34,34,35,36,
36,37,37,38,38,39,40,41,41,42,
43,44,45,45,46,47,48,48,49,50,
51,52,53,54,55,55,56,57,58,60,
61,63,64,65,66,68,70,71,72,74,
75,77,81,83,84,87,89,90,90,91]
```

```python
# graph size 
plt.figure(figsize=(15 , 10 )) # 15 is the width of graph  , 10 is the hight of graph 

# plotting the histogram 
plt.hist(age , bins= 10 ) ; 

#  bins are The bars of a histogram are called bins. The height of each bin shows how many values from that data fall into that range.
# The default value of the number of bins to be created in a histogram is 10
```

```python
# let's load dataset to understand 
import pandas as pd 
```

```python
data =pd.read_csv('StudentsPerformance.csv')
data.head()
```

```python
# graph size 
plt.figure(figsize=(15 , 10 )) # 15 is the width of graph  , 10 is the hight of graph 

plt.title("Distribution of math score for some students ")
# show how values of math score are distributed

sns.histplot(data['math score']) 
plt.show()
```

## Heatmap 

Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix. In this, to represent more common values or higher activities brighter colors basically reddish colors are used and to represent less common or activity values, darker colors are preferred. Heatmap is also defined by the name of the shading matrix. Heatmaps in Seaborn can be plotted by using the seaborn.heatmap() function.

```python
# We will be creating a 1010 2-D data using the randint() function of the NumPy module.
import numpy as np 
# generating 2-D 10x10 matrix of random numbers
# from 0 to 200
# low is the minimum number , high is the maximun number 
heat_graph = np.random.randint(size= (10 , 10 ) , low = 0  , high = 200  )
```

```python
# display data in matrix 
print(heat_graph)
```

```python
# graph size 
plt.figure(figsize=(15 , 10 )) # 10 is the width of graph  , y is the hight of graph 

# plotting the heatmap
sns.heatmap(data= heat_graph)

#If we want to display the value of the cells, then we pass the parameter annot as True. 
# sns.heatmap(data= heat_graph , annot= True)
```

### seaborn correlation heatmap

A correlation heatmap is a heatmap that shows a 2D correlation matrix between two discrete dimensions, using colored cells to represent data from usually a monochromatic scale . 
used to find the pairwise correlation of all columns in the dataframe. Any na values are automatically excluded.

```python
# graph size 
plt.figure(figsize=(10 , 7 )) # 10 is the width of graph  , 7 is the hight of graph 

# we have load the student performance data before , so we want to know the ocrrelations in this dataset 
# plotting the correlation heatmap using corr() function 
sns.heatmap(data.corr() , annot= True ) 

plt.show()
```

the correlation of a variable with itself is 1. For that reason all the diagonal values are 1.00

## boxplot 

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. 

Box plots are used to visualize summary statistics of a dataset, displaying attributes of the distribution like the datas range and distribution.



```python
# let's assume that we have list of values and we want to visualize it using box plot. 
my_sample = [20,10,2,34,54,12,3,5,6,8,13,15,17,19,9,18,22,11,1,32,25,19,30]  

# graph size 
plt.figure(figsize=(15 , 10 )) # 15 is the width of graph  , 10 is the hight of graph 

#plotting the boxplot 
plt.boxplot(my_sample) ; 
```

After creating the plot, we can see some of the summary statistics for our data. The Box Plot shows the median of the dataset (the vertical line in the middle), as well as the interquartile ranges (the ends of the boxes) and the minimum and maximum values of the dataset.

## Pie chart 

A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data. The area of slices of the pie represents the percentage of the parts of the data. The slices of pie are called wedges. 


```python
# let's assume that we have dataset about sales of some cars. 
name_of_car = ["BMW" , "TOYOTA" , "MG" , "NESSAN" , "HUNDAY"] 
sales_of_car = [ 1200 , 3000 ,2036 ,1598 ,2577 ] 


# graph size 
plt.figure(figsize=(15 , 10 )) # 15 is the width of graph  , 10 is the hight of graph 

#plotting the piechart 
plt.pie(sales_of_car , labels= name_of_car , autopct=  '%0.2f%%') ; 
# we use labels to identify what name of every preentage , but what is we want to show persentage of every car ? we use autopct attribute 

# you can use shadow attribute to makes shadow to pie 
```

```python
# graph size 
plt.figure(figsize=(15 , 10 )) # 15 is the width of graph  , 10 is the hight of graph 

#plotting the piechart with shadow 
# you can use shadow attribute to makes shadow to pie 
plt.pie(sales_of_car , labels= name_of_car , autopct=  '%0.2f%%' , shadow= True ) ; 
```

```python
# what if i want to explode of focuos on one part or the best part from cars ? 
# we use explode attribute 
# graph size 
plt.figure(figsize=(15 , 10 )) # 15 is the width of graph  , 10 is the hight of graph 

#plotting the piechart with shadow 
# you can use shadow attribute to makes shadow to pie 
plt.pie(sales_of_car , labels= name_of_car , autopct=  '%0.2f%%' , shadow= True  , explode = [0, 0.3 ,0 ,0 ,0 ]) ; 
```
